Question 1 : What is Python and why is it widely used in Data Analytics? Mention any three reasons.

- Answer 1

Python is a high-level, interpreted, general-purpose programming language known for its readability. It is the go-to for Data Analytics for three main reasons:
- 1 Extensive Libraries: Libraries like Pandas, NumPy, and Matplotlib make data manipulation and visualization effortless.

- 2 Simple Syntax: It’s easy to learn and read, allowing analysts to focus on solving data problems rather than complex code.

- 3 Large Community Support: Since so many people use it for data science, finding solutions to bugs or pre-built models is very easy.


Question 2 : Explain the difference between List and Tuple in Python.

- Answer 2

List

- Slightly slower due to dynamic nature.
- Mutable (can be changed after creation)
- Uses square brackets: [1, 2, 3]

Tuple

- Faster and more memory-efficient.
- Immutable (cannot be changed after creation)
- Uses parentheses: (1, 2, 3)



Question 3 : What is a function in Python? Why are functions useful?

- Answer 3

A function is a reusable block of code that performs a specific task.

Reusability: You write the code once and call it many times.

Organization: It breaks large, complex tasks into smaller, manageable pieces.

Readability: It makes the main program cleaner and easier to follow.

Question 10 : Write a Python program to count how many times a number appears in a list.

Example List

[2,3,4,2,5,2]

- Answer 10
